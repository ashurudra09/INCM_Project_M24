# Data visualization for the given datasets:
 
## 1) Iris Flower Dataset:

- Raw csv file can be found [here](../datasets/iris.csv)

```python
# Import necessary libraries
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# Load the iris dataset
iris_data = sns.load_dataset('iris')

# Display the first few rows of the dataset
print(iris_data.head())

# Pairplot to visualize relationships between numerical features in the iris dataset
sns.pairplot(iris_data, hue="species", diag_kind="kde", palette="bright")
plt.suptitle("Pairplot of Iris Dataset", y=1.02)
plt.show()

# Boxplot for sepal length across different species
plt.figure(figsize=(10, 6))
sns.boxplot(x='species', y='sepal_length', data=iris_data, palette="pastel")
plt.title('Sepal Length Distribution by Species')
plt.show()

# Violin plot for petal width categorized by species
plt.figure(figsize=(10, 6))
sns.violinplot(x='species', y='petal_width', data=iris_data, palette="muted")
plt.title('Petal Width Distribution by Species')
plt.show()
```

## 2) Penguins Dataset:

- Raw csv file can be found [here](../datasets/penguins.csv).

```python

# Import necessary libraries
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# Load the penguin dataset
penguin_data = sns.load_dataset('penguins')

# Handle missing values by dropping rows with NaN values
penguin_data = penguin_data.dropna()

# Display the first few rows of the dataset
print(penguin_data.head())

# Pairplot to visualize relationships between numerical features in the penguin dataset
sns.pairplot(penguin_data, hue="species", diag_kind="kde", palette="bright")
plt.suptitle("Pairplot of Penguin Dataset", y=1.02)
plt.show()

# Boxplot for bill length across different islands
plt.figure(figsize=(10, 6))
sns.boxplot(x='island', y='bill_length_mm', hue='species', data=penguin_data)
plt.title('Bill Length Distribution by Island and Species')
plt.show()

# Violin plot for body mass categorized by sex
plt.figure(figsize=(10, 6))
sns.violinplot(x='sex', y='body_mass_g', hue='species', data=penguin_data, split=True, palette="muted")
plt.title('Body Mass Distribution by Sex and Species')
plt.show()
```
