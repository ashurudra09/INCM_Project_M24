# Data visualization for the given datasets:
 
## 1) Iris Flower Dataset:

- Raw csv file can be found [here](../datasets/iris.csv)

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv("../datasets/iris.csv")

print("General Statistics:")
print("A) Means by Species:")
print(df.groupby("species").mean().head())
print("B) Standard Deviations by Species:")
print(df.groupby("species").agg(np.std, ddof=0).head())
```

## Sepals:
### Distributions:

```python
sl_setosa = df["sepal_length"][(df["species"] == "setosa")]
sl_versicolor = df["sepal_length"][(df["species"] == "versicolor")]
sl_virginica = df["sepal_length"][(df["species"] == "virginica")]

# Creating the histogram
plt.figure(figsize=(10, 6))
plt.hist(sl_setosa, bins=10, color="blue", alpha=0.5, label="Setosa")
plt.hist(sl_versicolor, bins=10, color="green", alpha=0.5, label="Versicolor")
plt.hist(sl_virginica, bins=10, color="red", alpha=0.5, label="Virginica")

# Adding labels and title
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Frequency")
plt.title("Histogram of Sepal Length by Species")
plt.legend()
plt.show()
```

```python
sw_setosa = df["sepal_width"][(df["species"] == "setosa")]
sw_versicolor = df["sepal_width"][(df["species"] == "versicolor")]
sw_virginica = df["sepal_width"][(df["species"] == "virginica")]

# Creating the histogram
plt.figure(figsize=(10, 6))
plt.hist(sw_setosa, bins=10, color="blue", alpha=0.5, label="Setosa")
plt.hist(sw_versicolor, bins=10, color="green", alpha=0.5, label="Versicolor")
plt.hist(sw_virginica, bins=10, color="red", alpha=0.5, label="Virginica")

# Adding labels and title
plt.xlabel("Sepal Width (cm)")
plt.ylabel("Frequency")
plt.title("Histogram of Sepal Width by Species")
plt.legend()
plt.show()
```

### Sepal Length vs Sepal Width:

```python
# Create scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(sl_setosa, sw_setosa, color="blue", marker="o", label="Setosa")
plt.scatter(sl_versicolor, sw_versicolor, color="green", marker="s", label="Versicolor")
plt.scatter(sl_virginica, sw_virginica, color="red", marker="^", label="Virginica")

# Labels and legend
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Sepal Width (cm)")
plt.title("Sepal Length vs Sepal Width by Species")
plt.legend()
plt.show()
```

## Petals:
### Distributions:

```python
pl_setosa = df["petal_length"][df["species"] == "setosa"]
pl_versicolor = df["petal_length"][df["species"] == "versicolor"]
pl_virginica = df["petal_length"][df["species"] == "virginica"]

# Creating the histogram
plt.figure(figsize=(10, 6))
plt.hist(pl_setosa, bins=10, color="blue", alpha=0.5, label="Setosa")
plt.hist(pl_versicolor, bins=10, color="green", alpha=0.5, label="Versicolor")
plt.hist(pl_virginica, bins=10, color="red", alpha=0.5, label="Virginica")

# Adding labels and title
plt.xlabel("Petal Length (cm)")
plt.ylabel("Frequency")
plt.title("Histogram of Petal Length by Species")
plt.legend()
plt.show()
```

```python
pw_setosa = df["petal_width"][df["species"] == "setosa"]
pw_versicolor = df["petal_width"][df["species"] == "versicolor"]
pw_virginica = df["petal_width"][df["species"] == "virginica"]

# Creating the histogram
plt.figure(figsize=(10, 6))
plt.hist(pw_setosa, bins=10, color="blue", alpha=0.5, label="Setosa")
plt.hist(pw_versicolor, bins=10, color="green", alpha=0.5, label="Versicolor")
plt.hist(pw_virginica, bins=10, color="red", alpha=0.5, label="Virginica")

# Adding labels and title
plt.xlabel("Petal Width (cm)")
plt.ylabel("Frequency")
plt.title("Histogram of Petal Width by Species")
plt.legend()
plt.show()
```

### Petal Length vs Petal Width:
